# **Line Plot â€“ Temperature Trend**
Draw a line plot for the average daily temperature of a city for one month (30 days). Identify 
the hottest and coldest days.

```python
import matplotlib.pyplot as plt

data = {
    1: 22,  2: 24,  3: 21,  4: 23,  5: 25,
    6: 27,  7: 28,  8: 26,  9: 30, 10: 29,
    11: 31, 12: 32, 13: 28, 14: 27, 15: 26,
    16: 24, 17: 22, 18: 23, 19: 25, 20: 27,
    21: 29, 22: 30, 23: 28, 24: 26, 25: 24,
    26: 23, 27: 22, 28: 21, 29: 23, 30: 25
}

days = list(data.keys())
temp = list(data.values())
hottest = max(data, key=data.get)
coldest = min(data, key=data.get)
hottest_temp = data[hottest]
coldest_temp = data[coldest]
plt.figure(figsize=(10, 5))
plt.plot(days, temp, marker='o', linestyle='--', color='b', label="Temperature")
plt.scatter(hottest, hottest_temp, color='red', s=100, label=f"Hottest Day ({hottest_temp}Â°C)")
plt.scatter(coldest, coldest_temp, color='cyan', s=100, label=f"Coldest Day ({coldest_temp}Â°C)")
plt.title("Average Temperature(past 30 Days)")
plt.xlabel("Day of Month")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)

plt.show()

```

# **Line Plot â€“ Stock Prices**
Plot the closing stock prices of a company over 15 days using a line plot. Highlight the 
maximum and minimum prices with markers.

```python
import random
stock = {day: random.randint(13000, 16000) for day in range(1, 16)}
days_stock = list(stock.keys())
prices = list(stock.values())
max_price_day = max(stock, key=stock.get)
min_price_day = min(stock, key=stock.get)
max_price = stock[max_price_day]
min_price = stock[min_price_day]

plt.figure(figsize=(10, 5))
plt.plot(days_stock, prices, marker='o', linestyle='-', color='g', label="Stock Prices")
plt.scatter(max_price_day, max_price, color='red', s=100, label=f"Max Price (Rs.{max_price})")
plt.scatter(min_price_day, min_price, color='blue', s=100, label=f"Min Price (Rs.{min_price})")
plt.title("Stock Prices Over 15 Days")
plt.xlabel("Day")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

# **Multiple Line Plot â€“ Sales Comparison**
Create a multiple line plot comparing monthly sales of two products (Product A and Product 
B) for one year.

```python
import random

months = list(range(1, 13))
product_a_sales = [random.randint(200, 400) for _ in range(12)]
product_b_sales = [random.randint(200, 400) for _ in range(12)]

plt.figure(figsize=(10, 5))
plt.plot(months, product_a_sales, marker='o', linestyle='-', color='b', label="Product A Sales")
plt.plot(months, product_b_sales, marker='s', linestyle='--', color='r', label="Product B Sales")

plt.title("Monthly Sales Comparison")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(months)
plt.legend()
plt.grid(True)
plt.show()

```

# **Histogram â€“ Exam Scores**
Given exam scores of 100 students (range 0â€“100), create a histogram to show the frequency 
distribution. Identify the most common score range.

```python

import seaborn as sns
import numpy as np
exam_scores = [random.randint(0, 100) for _ in range(100)]

plt.figure(figsize=(10, 5))
sns.histplot(exam_scores, bins=10, kde=True, color='blue')

plt.title("Exam Scores Distribution with KDE")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

counts, bins = np.histogram(exam_scores, bins=10)
most_common_range = f"{int(bins[counts.argmax()])}â€“{int(bins[counts.argmax() + 1])}"
print(f"The most common score range is: {most_common_range}")
```

# **Histogram â€“ Daily Steps**
A person tracked their daily steps for 30 days. Create a histogram to analyze step 
distribution. Comment on whether the person is consistent

```python
steps = [random.randint(2000, 10000) for _ in range(30)]
plt.figure(figsize=(10, 5))
sns.histplot(steps, bins=30, kde=True, color='green')

plt.title("Daily Steps Distribution")
plt.xlabel("Steps Range")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

if max(steps) > sum(steps) / len(steps):
    print("The person is consistent in their daily steps.")
else:
    print("The person is not consistent in their daily steps.")
```

# **Combined Visualization â€“ Sales Data**
For monthly shop sales: Jan: 200, Feb: 250, Mar: 300, Apr: 280, May: 350, Jun: 400: 
â€¢ Create a bar plot of sales. 
â€¢ On the same figure, plot a line plot showing the sales trend. 
â€¢ Also, prepare a histogram of sales values to analyze sales frequency distribution

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [200, 250, 300, 280, 350, 400]

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.bar(months, sales, color='skyblue', label="Monthly Sales")
plt.plot(months, sales, marker='o', color='red', label="Sales Trend")
plt.title("Monthly Sales and Trend")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.subplot(1, 2, 2)
sns.histplot(sales, bins=5, kde=True, color='green')
plt.title("Sales Frequency Distribution")
plt.xlabel("Sales Range")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
