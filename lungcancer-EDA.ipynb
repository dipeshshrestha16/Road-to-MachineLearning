```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Loading the dataset**

```python
file_path = 'D:\pythonML-class\survey lung cancer.csv'
df = pd.read_csv(file_path)

print(df.info())
print(df.head())
```

# **Data Cleaning**

```python
# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Standardize categorical values
df['GENDER'] = df['GENDER'].str.upper().str.strip()
df['LUNG_CANCER'] = df['LUNG_CANCER'].str.upper().str.strip()

# Check unique values
print("\nUnique GENDER:", df['GENDER'].unique())
print("Unique LUNG_CANCER:", df['LUNG_CANCER'].unique())
```

```python
df.head()
```

# **Explarotary Data Analysis**

### **Age distribution**

```python
plt.figure(figsize=(8,5))
sns.histplot(df['AGE'], bins=20, kde=True, color='blue')
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

### **Gender Distribution**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='GENDER', data=df, palette='Set2')
plt.title("Gender Distribution")
plt.show()
```

### **Lung Cancer Diagnosis Count**

```python
df['LUNG_CANCER'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['skyblue', 'salmon'])
sns.countplot(x='LUNG_CANCER', data=df, palette='Set1')
plt.title("Lung Cancer Diagnosis Count")
plt.show()
```

### **Smoking Vs Lung Cancer**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='SMOKING', hue='LUNG_CANCER', data=df, palette='viridis')
plt.title("Smoking vs Lung Cancer")
plt.show()
```

### **Heatmap**

```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap of Lung Cancer Dataset")
plt.show()
```

### **Exploring the dataset in more detail**

```python
plt.figure(figsize=(4,4))
sns.countplot(data=df,x='COUGHING',hue='LUNG_CANCER')
plt.legend(["Has cancer", 'Does not have cancer'])
```
