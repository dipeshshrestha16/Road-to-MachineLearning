```python
#reading csv file
import pandas as pd
import seaborn as sns
import plotly.express as px

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"]=(10,6)

df=pd.read_csv(r"D:\pythonML-class\netflix_titles_sample.csv")
print("Shape of dataset:",df.shape)
print(df.head())
```

```python
df['director'].fillna("Unknown", inplace=True)
df['Cast'].fillna("Unknown",inplace=True)
df['country'].fillna("Unknown",inplace=True)
df['rating'].fillna("Unknown",inplace=True)
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce')
df['date_added']=pd.fillna(pd.Timestamp("1900-01-01"))

```

```python
sns.countplot(x='type', data=df, palette="Set2")
plt.title("Movies")
plt.show()
```

```python
top_countries=df['country'].value_counts().nlargest(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title("top 10 content")
plt.xlabel("No of titles")
plt.ylabel("Countries")
plt.show()
```

```python
fig=px.histogram(df,x="release_year",color="type",barmode="group",title="Movies vs TV shows")
fig.show()
```

```python
#reading csv file
import pandas as pd
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"]=(10,6)

df=pd.read_csv(r"D:\pythonML-class\netflix_titles_sample.csv")
print("Shape of dataset:",df.shape)
print(df.head())
```

# **distribution of rating**

```python
# Distribution of rating
sns.countplot(x='rating', data=df, order=df['rating'].value_counts().index, palette="coolwarm")
plt.title("Distribution of Ratings")
plt.show()
```

# **titles added per year**


```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
sns.countplot(x='year_added', data=df, palette="viridis")
plt.title("Titles Added Per Year")
plt.show()
```

# **top 10 longest movies**

```python
movies = df[df['type'] == 'Movie'].copy()
# Extract duration for numbers and intigers
movies['duration_int'] = movies['duration'].str.split().str[0].astype(float)
top10 = movies.nlargest(10, 'duration_int')
sns.barplot(x='title', y='duration_int', data=top10, palette="crest")
plt.title("Top 10 Longest Movies")
plt.xlabel("Movie Title")
plt.ylabel("Duration (minutes)")
plt.show()
```

# **interactive visualization with plotly**

```python

fig=px.histogram(df,x="release_year",color="type",barmode="group",title="Movies vs TV shows")
fig.show()
```

# **pairplot**

```python

sns.pairplot(movies[['release_year', 'duration_int']].dropna())
plt.suptitle("Pairplot: Release Year vs Duration (Movies)")
plt.show()
```
