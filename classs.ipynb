```python
import pandas as pd
data_pivot={
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Department": ["IT", "HR", "HR", "IT"],
    "Salary": [50000, 60000, 55000, 52000],
    "Gender": ["F", "M", "F", "M"]

}
df_pivot=pd.DataFrame(data_pivot)
pd.pivot_table(df_pivot, values="Salary", index="Department", columns="Gender", aggfunc="mean")
```

```python
import numpy as np
data=[7, 8, 9, 10, 11, 12, 13, 14, 15]

data.sort()

#calculating Q1, Q2, Q3
Q1=np.percentile(data, 25)
Q2=np.percentile(data, 50)
Q3=np.percentile(data, 75)
IQR=Q3-Q1

print("Data:", data)
print("Q1(25th percentile):", Q1)
print("Q2(median/50th percentile):", Q2)
print("Q3(75th percentile):", Q3)
print("IQR(Interquartile Range):", IQR)
```

```python
import matplotlib.pyplot as plt

data=[7, 8, 3, 10, 5, 7, 13, 12, 15]

plt.figure(figsize=(20,5))
plt.plot(data, marker='o', linestyle='--', color='b',linewidth=2, markersize=8)
plt.title('Line Plot Example', fontsize=36)
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.xticks([0,1,2,3,4,5,6,7,8],['A','B','C','D','E','F','G','H','I'])
plt.yticks([0,2,4,6,8,10,12,14,16],[0,2,4,6,8,10,12,14,16])
plt.ylim(0,16)

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(range(len(data)),data, color='skyblue')
plt.title('Bar Chart Example', fontsize=20)
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(axis='y')
plt.xticks(range(len(data)),['A','B','C','D','E','F','G','H','I'])


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data=[7, 8, 3, 10, 5, 7, 13,3, 12, 7, 15, 13,7,8]
plt.figure(figsize=(10,5))
sns.histplot(data, bins=8, kde=True, color='purple')
plt.title('Histogram Example', fontsize=20)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

```
